```python
plt.hlines(exact,0,300)
plt.plot(convergence_bf_095)
plt.plot(convergence_bf_098)
plt.plot(convergence_bf_10)
plt.plot(convergence_bf_105)


# plt.plot(convergence_gs_048)
# plt.plot(convergence_gs_05)
# plt.plot(convergence_gs_052)
plt.xlabel("Iteration")
plt.ylabel("Energy")
plt.title("Energy using Metropolis Brute force as a function of gradient descent iteration for an interacting system of 2 particle in 2 dimensions \n with 500000 mc-iterations with optimal parameters and varying $\sigma^2$", fontsize=15)
# plt.plot(convergence_gs_052)
plt.legend([r"$\sigma^2=0.95$", r"$\sigma^2=0.98$", r"$\sigma^2 = 1.0$",  r"$\sigma^2 = 1.05$"])
plt.ylim(2.99, 3.15)
plt.savefig("plots/interacting_num_particles_2_num_dims_2_bf_varying_sigma.png")
plt.show()
```
